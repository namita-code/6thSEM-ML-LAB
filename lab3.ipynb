```python
import numpy as np
import pandas as pd

# Load dataset
data = pd.read_csv("/content/canada_per_capita.csv")

# Select columns correctly
X = data[['year']].values        # Convert to numpy array
y = data[['per capita income (US$)']].values

# Add bias term (column of 1s)
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))

# Normal Equation
XT = X.T
theta = np.linalg.inv(XT @ X) @ XT @ y

print("Intercept (θ0):", theta[0][0])
print("Slope (θ1):", theta[1][0])

# Prediction
year = 2020
prediction = theta[0][0] + theta[1][0] * year
print("Predicted value for", year, ":", prediction)
```

```python
import numpy as np
import pandas as pd

# Load dataset
data = pd.read_csv("/content/salary.csv")

# Handle missing values
data.fillna(data.mean(), inplace=True)

# Select input and output
X = data[['YearsExperience']].values   # Convert to numpy array
y = data[['Salary']].values            # Make y 2D column vector

# Add bias term (column of 1s)
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))

# Normal Equation
XT = X.T
theta = np.linalg.inv(XT @ X) @ XT @ y

print("Intercept (θ0):", theta[0][0])
print("Slope (θ1):", theta[1][0])

# Prediction
experience = 12
prediction = theta[0][0] + theta[1][0] * experience
print("Predicted salary for", experience, "years:", prediction)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


data = pd.read_csv("/content/canada_per_capita.csv")


X = data[['year']]
y = data['per capita income (US$)']

model = LinearRegression()

model.fit(X, y)


print("Intercept (θ0):", model.intercept_)
print("Slope (θ1):", model.coef_[0])


year = 2020
prediction = model.predict([[year]])
print("Predicted value for", year, ":", prediction[0])
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("hiring.csv")

# Convert words to numbers
df['experience'] = df['experience'].replace({
    'two': 2,
    'three': 3,
    'five': 5,
    'seven': 7,
    'ten': 10,
    'eleven': 11
})

df['experience'] = df['experience'].fillna(0)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())

# Feature matrix
X = df[['experience','test_score(out of 10)','interview_score(out of 10)']].values
Y = df['salary($)'].values

# Add column of 1s for intercept
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))

# Matrix formula
Xt = X.T
beta = np.linalg.inv(Xt.dot(X)).dot(Xt).dot(Y)

print("Beta values (Intercept, b1, b2, b3):")
print(beta)

# Prediction example (2 yrs exp, 9 test, 6 interview)
new_data = np.array([1, 2, 9, 6])
prediction = new_data.dot(beta)

print("Predicted salary:", prediction)
```
